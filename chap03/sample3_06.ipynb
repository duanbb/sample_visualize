```python
# サンプル3-06-1
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# 0～14の範囲で100個の等差数列を生成
x = np.linspace(0, 14, 100)
# x + i*0.5の正弦（サイン）をプロット
for i in range(5):
    plt.plot(x, np.sin(x + i*0.5))

plt.show()
```

```python
# サンプル3-06-2
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns # Seabornのインポート

sns.set()             # Seabornで出力できるようにする
x = np.linspace(0, 14, 100)
for i in range(5):
    plt.plot(x, np.sin(x + i*0.5))

plt.show()            # Seabornのグラフとして出力される
```
