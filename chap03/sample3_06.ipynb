```python
# 3-06-1
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# 014100
x = np.linspace(0, 14, 100)
# x + i*0.5
for i in range(5):
    plt.plot(x, np.sin(x + i*0.5))

plt.show()
```

```python
# 3-06-2
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns # Seaborn

sns.set()             # Seaborn
x = np.linspace(0, 14, 100)
for i in range(5):
    plt.plot(x, np.sin(x + i*0.5))

plt.show()            # Seaborn
```
