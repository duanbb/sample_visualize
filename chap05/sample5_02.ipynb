```python
# サンプル5-2-1
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.set(style='darkgrid')    # グリッドを表示
df = pd.DataFrame(
    dict(
        # timeの値として0から500までの等差数列
        time=np.arange(500),
        # valueの値として標準正規分布から500個抽出し、累積合計を求める
        value=np.random.randn(500).cumsum()
        )
    )
fg = sns.relplot(x='time',    # x値
                 y='value',   # y値
                 kind='line', # ラインプロット
                 data=df      # データフレーム
                 )
fg.fig.autofmt_xdate()        # x軸ラベルを30度傾ける
plt.show()
```

```python
# サンプル5-2-2
fmri = sns.load_dataset('fmri') # fMRI(磁気共鳴機能画像法)のデータ
print(fmri)
```

```python
# サンプル5-2-3
sns.relplot(x='timepoint', # xは時系列のポイント
            y='signal',    # yは信号の値
            hue='subject', # 被検者別にラインプロット
            col='region',  # 横方向に検査領域(頭頂壁,正面)
            row='event',   # 縦方向に刺激,合図の別
            height=6,      # 高さを5インチ(幅も6になる)
            kind='line',   # ラインプロット
            data=fmri
           )
plt.show()
```
