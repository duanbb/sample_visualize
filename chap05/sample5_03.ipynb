```python
# サンプル5-3-1
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips') # tipsデータセット
```

```python
sns.catplot(x='day',        # x値
            y='total_bill', # y値
            data=tips       # データフレーム
           )
plt.show()
```

```python
# サンプル5-3-2
sns.catplot(x='day',
            y='total_bill',
            kind='swarm',   # swarmplot()を使う
            data=tips
           )
plt.show()
```

```python
# サンプル5-3-3
sns.catplot(x='day',        # x値
            y='total_bill', # y値
            hue='sex',      # 男女の別で色分けする
            kind='swarm',   # swarmplot()を使う
            data=tips
           )
plt.show()
```

```python
# サンプル5-3-4
sns.catplot(x='total_bill', # x軸に変量を指定
            y='day',        # y軸にカテゴリ変数を指定
            hue='time',     # ランチ,ディナーで色分け
            kind='swarm',   # swarmplot()を使う
            data=tips
           )
plt.show()
```

```python
# サンプル5-3-5
sns.catplot(x='day',
            y='total_bill',
            kind='box',     # ボックスプロット
            data=tips
           )
plt.show()
```

```python
# サンプル5-3-6
sns.catplot(x="day",
            y="total_bill",
            hue="smoker",   # 喫煙の有無を加える
            kind="box",     # ボックスプロット
            data=tips
           )
plt.show()
```

```python
# サンプル5-3-7
sns.catplot(x='day',          # x軸のカテゴリは曜日
            y='total_bill',   # 総支払額の分布をプロット
            hue='sex',        # 男女の別を加える
            kind='violin',    # バイオリンプロットにする
            inner='quartile', # 分布の四分位数を表示
            split=True,       # 本体を2つに分割
            data=tips
           )
plt.show()
```

```python
# サンプル5-3-8
titanic = sns.load_dataset('titanic')
sns.catplot(x='deck',         # デッキごとの人数
            hue='class',
            kind='count',     # カウントプロット
            palette='pastel', # カラーパレットpastelを指定
            data=titanic
           )
plt.show()
```
