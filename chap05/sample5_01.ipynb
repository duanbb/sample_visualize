```python
# サンプル5-1-1
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# サンプル5-1-2
tips = sns.load_dataset('tips') # tipsデータセット
print(tips)
```

```python
# サンプル5-1-3
sns.relplot(x='total_bill', # x値
            y='tip',        # y値
            data=tips       # データフレーム
           )
plt.show()
```

```python
# サンプル5-1-4
sns.relplot(x='total_bill', # x値
            y='tip',        # y値
            hue='sex',      # 第3の変量で色分けする
            style='sex',    # 第3の変量でマーカーの形状を変える
            data=tips
           )
plt.show()
```

```python
# サンプル5-1-5
sns.relplot(x='total_bill', # x値
            y='tip',        # y値
            hue='sex',      # sexで色分けする
            style='smoker', # smokerでマーカーの形状を変える
            data=tips
           )
plt.show()
```

```python
# サンプル5-1-6
sns.relplot(x='total_bill', # x値
            y='tip',        # y値
            hue='size',     # sizeの0～6の値で色分けする
            data=tips
           )
plt.show()
```

```python
# サンプル5-1-7
sns.relplot(x='total_bill', # x値
            y='tip',        # y値
            size='size',    # size列の値の大きさによってマーカーサイズを変える
            data=tips
           )
plt.show()
```

```python
# サンプル5-1-8
sns.set(style='darkgrid')    # グリッドを表示
sns.relplot(x='total_bill',  # x値
            y='tip',         # y値
            size='size',     # size列の値の大きさによってマーカーサイズを変える
            sizes=(15, 220), # マーカーサイズの最小と最大サイズを指定
            data=tips
           )
plt.show()
```

```python
# サンプル5-1-9
sns.set(style="darkgrid")   # グリッドを表示
sns.relplot(x="total_bill", # x値
            y="tip",        # y値
            hue="sex",      # 男女の別で色分けする
            col="time",     # ランチ、ディナーでそれぞれ散布図を作成
            data=tips
           )
plt.show()
```

```python
# サンプル5-1-10
sns.set(style="darkgrid")   # グリッドを表示
sns.relplot(x="total_bill", # x値
            y="tip",        # y値
            hue="sex",      # 男女の別で色分けする
            col="day",      # ヨコ方向は曜日別
            row='time',     # タテ方向はランチ、ディナーの別
            data=tips
           )
plt.show()
```
