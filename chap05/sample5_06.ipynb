```python
# サンプル5-05-1
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set(style='darkgrid')       # グリッドを表示
tips = sns.load_dataset("tips") # tipsデータセットを読み込む
sns.regplot(x="total_bill",     # x軸のデータは総支払額
            y="tip",            # y軸のデータはチップの額
            data=tips
           )
```
