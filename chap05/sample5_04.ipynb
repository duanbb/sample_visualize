```python
# サンプル5-04-1
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='darkgrid')      # グリッドを表示
x = np.random.normal(size=100) # 標準正規分から100個のデータをサンプリング
```

```python
# サンプル5-04-2
sns.distplot(x)
plt.show()
```

```python
# サンプル5-04-3
sns.distplot(x, kde=False, norm_hist=False)
plt.show()
```

```python
# サンプル5-04-4
# 平均、共分散行列を設定
mean, cov = [0, 1], [(1, .5), (.5, 1)]
# 乱数を300×2作成して300行×2列の行列にする
data = np.random.multivariate_normal(mean, cov, 300)
# 行列の各列をx、yとするデータフレームを作成
df = pd.DataFrame(data, columns=['x', 'y'])
# 二変量グラフと単変量グラフを描く
sns.jointplot(x="x", y="y", data=df)
```
