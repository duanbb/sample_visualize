```python
# サンプル4-02-1
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

font = {'family': 'serif',             # フォントファミリー
        'color' : 'navy',              # フォントカラー
        'style' : 'italic',            # イタリック
        'weight': 'bold',              # フォントウェイト
        'size'  : 6,                   # フォントサイズ
        'rotation': 90                 # 90度回転
        }

# 0～14の範囲で100個の等差数列を生成
x = np.linspace(0, 14, 100)
plt.plot(x, np.sin(x))                 # xの正弦をプロット
plt.title('MaineTitle', fontdict=font) # メインタイトル
plt.xlabel('x-label', fontdict=font)   # x軸ラベル
plt.ylabel('y-label', fontdict=font)   # y軸ラベル

plt.show()
```
