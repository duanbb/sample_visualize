```python
# 4-38-1
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

def func(x,                  # int: f(x,y)x
         y                   # int: f(x,y)y
         ):
    '''
    3

    '''
    return x**2 + y**2       # float: f(x, y)=x^2 + y^2


x1 = np.arange(-3, 3, 0.25)  # x_1
x2 = np.arange(-3, 3, 0.25)  # x_2
X, Y = np.meshgrid(x1, x2)   # 2
Z = func(X, Y)               # f(x, y)X,Y-3
                             # 30.25Z

fig = plt.figure()           # Figure
ax = axes3d.Axes3D(fig)      # Axes3D
ax.set_xlabel("x1")          # x1
ax.set_ylabel("x2")          # x2
ax.set_zlabel(r"$f_(x, y)$", size=15)    # f(x1,x2)
ax.plot_wireframe(X,Y,Z)     # x1x2 f(x1,x2)

plt.show()                   # 
```
