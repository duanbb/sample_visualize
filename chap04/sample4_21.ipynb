```python
# サンプル4-20-1
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0, 2.0, 0.001)# 0.0から2.0まで0.001刻みの等差数列
y = np.sin(4 * np.pi * x)     # 4π・ｘの正弦

upper = 0.75                  # 上側の境界
lower = -0.75                 # 下側の境界

# 上側のライン
set_upper = np.ma.masked_where(y < upper, # 上側境界値より小さい値をマスクする
                               y)
# 下側のライン
set_lower = np.ma.masked_where(y > lower, # 下側境界値より大きい値をマスクする
                               y)
# 中間のライン
set_middle = np.ma.masked_where(        
    np.logical_or(y > upper,              # 上側境界よりも大きい値をマスクする
                  y < lower), y)          # 下側境界よりも小さい値をマスクする

fig, ax = plt.subplots()
ax.plot(x, set_upper, color='red')        # 上側のラインをプロット
ax.plot(x, set_lower, color='blue')       # 下側のラインをプロット
ax.plot(x, set_middle, color='black')     # 中間のラインをプロット

plt.show()
```
