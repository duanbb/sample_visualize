```python
# サンプル4-23-2
fig, ax = plt.subplots(figsize=(2, 8))

# アクセント
ax.text(0.1,0.9, r'$\acute x$', size=20)        # アキュート
ax.text(0.1,0.8, r'$\bar x$',   size=20)        # バー
ax.text(0.1,0.7, r'$\brevex$',  size=20)        # ブリーブ
ax.text(0.1,0.6, r'$\ddot x$',  size=20)        # ダブルドット
ax.text(0.1,0.5, r'$\dot x$',   size=20)        # ドット
ax.text(0.1,0.4, r'$\grave x$', size=20)        # グレーブ
ax.text(0.1,0.3, r'$\hat x$',   size=20)        # ハット
ax.text(0.1,0.2, r'$\tilde x$', size=20)        # チルダ
ax.text(0.1,0.1, r'$\vec x$',   size=20)        # 右向き矢印
ax.text(0.1,0.0, r'$\overline {xyz}$', size=20) # ライン

plt.show()
```

```python
# サンプル4-23-3
fig, ax = plt.subplots(figsize=(3, 8))

# ルート記号
ax.text(0.1,0.9, r'$\sqrt{2}$', size=15)     # √2
ax.text(0.1,0.8, r'$\sqrt[3]{x}$', size=15)  # 3√x

# 分数
ax.text(0.1,0.7, r'$\frac{3}{4} - \frac{3}{4} - \frac{3}{4}$', size=13)
# 分数、(横棒なしカッコ付き)、横棒なし
ax.text(0.1,0.6, r'$\frac{3}{4} \binom{3}{4} \stackrel{3}{4}$', size=13)
# 分数のネスト
ax.text(0.1,0.5, r'$\frac{5 - \frac{1}{x}}{4}$', size=15)
# 全体を( )で囲むとカッコの大きさが小さくなる
ax.text(0.1,0.4, r'$(\frac{5 - \frac{1}{x}}{4})$', size=15)
# カッコの前に\leftと\right を付けると適正なサイズになる
ax.text(0.1,0.3, r'$\left(\frac{5 - \frac{1}{x}}{4}\right)$', size=15)

# 下付き文字は_、上付きは^を使用
ax.text(0.1,0.2, r'$\alpha_i + \beta^i$', size=15)

# 総和記号の例
ax.text(0.1,0.1, r'$\sum_{i=0}^\infty x_i$', size=20)

plt.show()
```
