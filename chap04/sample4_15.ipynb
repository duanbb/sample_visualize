```python
# サンプル4-15-1
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 400) # 0～π×2を400等分した等差数列
y = np.sin(x ** 2)                 # 値を2乗して正弦(sin)を求める
```

```python
# サンプル4-15-2
f, ax = plt.subplots(3, 1,       # 3×1のグリッド
                     sharex=True # x軸を共有
                     )
f.suptitle('Sharing x-axes')     # Figureのタイトル
ax[0].plot(x, y)                 # 1行目のAxesにx**2の正弦曲線をプロット
ax[1].scatter(x, y, s=10)        # 2行目のAxesにx**2の正弦の分布をプロット
ax[2].scatter(x, 2*y** 2 - 1,    # 3行目のAxesに「2*y**2 - 1」の分布をプロット
              s=10,
              color='r'
              )

plt.show()
```

```python
# サンプル4-15-3
f, ax = plt.subplots(3, 1,       # 3×1のグリッド
                     sharex=True # x軸を共有
                     )
f.suptitle('Sharing x-axes')     # Figureのタイトル
ax[0].plot(x, y)                 # 1行目のAxesにx**2の正弦曲線をプロット
ax[1].scatter(x, y, s=10)        # 2行目のAxesにx**2の正弦の分布をプロット
ax[2].scatter(x, 2 * y ** 2 - 1, # 3行目のAxesに「2*y**2 - 1」の分布をプロット
              s=10,
              color='r'
              )
f.subplots_adjust(hspace=0)          # 縦方向のスペースを0にする

plt.show()
```

```python
# サンプル4-15-4
f, ax = plt.subplots(1, 3, sharey=True) # 1×3のグリッド
f.suptitle('Sharing y-axes')
ax[0].plot(x, y)                 # 1行目のAxesにxの**2の正弦曲線をプロット
ax[1].scatter(x, y, s=10)        # 2行目のAxesにx**2の正弦の分布をプロット
ax[2].scatter(x, 2 * y ** 2 - 1, # 3行目のAxesに「2*y**2 - 1」の分布をプロット
              s=10, color='r')
plt.show()
```

```python
# サンプル4-15-5
f, ax = plt.subplots(1, 3, sharey=True) # 1×3のグリッド
f.suptitle('Sharing y-axes')
ax[0].plot(x, y)                 # 1行目のAxesにx**2の正弦曲線をプロット
ax[1].scatter(x, y, s=10)        # 2行目のAxesにx**2の正弦の分布をプロット
ax[2].scatter(x, 2 * y ** 2 - 1, # 3行目のAxesに「2*y**2 - 1」の分布をプロット
              s=10, color='r')
f.subplots_adjust(wspace=0)      # 横方向のスペースを0にする

plt.show()
```
