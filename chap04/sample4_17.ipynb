```python
# サンプル4-17-1
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

num = np.linspace(0, 2 * np.pi, 100) # 0～π・2を100等分
x = 3 * np.cos(num)                  # 等差数列の余弦(cos)
y = 3 * np.sin(num)                  # 等差数列の正弦(sin)
```

```python
# サンプル4-17-2
fig, axs = plt.subplots() # 初期設定の状態でFigure、Axesを生成
axs.plot(x, y)            # xを余弦、yを正弦として曲線をプロット
plt.show()
```

```python
# サンプル4-17-3
fig, axs = plt.subplots(figsize=(6, 6)) # Figureの縦横比を同じにする
axs.plot(x, y)
plt.show()
```
