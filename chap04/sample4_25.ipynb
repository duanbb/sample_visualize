```python
# サンプル4-25-1
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

plt.rcdefaults()         # グラフのスタイルのデフォルト設定を読み込む
fig, ax = plt.subplots() # FigureにAxesを配置

# 乱数生成のシード（種）
np.random.seed(0)
# データの作成
fruit = ('Apple', 'Orange', 'Grape', 'Banana', 'Pineapple')
y_pos = np.arange(len(fruit)) # 0からfruitの数だけ等差数列を生成
# 0.0以上1.0未満の範囲からfruitの要素の数だけ乱数を生成し20を掛ける
performance = 20 * np.random.rand(len(fruit))
# 誤差として0.0以上1.0未満の範囲からfruitの要素の数だけ乱数を生成
error = np.random.rand(len(fruit))

# 水平バーの描画
ax.barh(y_pos,               # 各水平バーのy座標
        performance,         # バーの幅（長さ）
        xerr=error,          # 水平のエラーバーを表示
        align='center',      # バーをy軸の中央に表示
        color='pink',        # バーの色
        ecolor='black'       # エラーバーの色  
       )
ax.set_yticks(y_pos)         # y軸の目盛ラベルを棒ごとに表示
ax.set_yticklabels(fruit)    # 棒ごとのラベル
ax.invert_yaxis()            # y軸を反転させて最上部をAppleにする
ax.set_xlabel('Performance') # x軸ラベル
ax.set_title("Today's sales forecast") # グラフタイトル

plt.show()
```
