```python
# サンプル4-16-1
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0, 2 * np.pi, 400) # 0～π×2を400等分した等差数列
y1 = np.sin(x1 ** 2)                # 値を2乗して正弦(sin)を求める
x2 = np.linspace(0, 4 * np.pi, 400) # 0～π×4を400等分した等差数列
y2 = np.sin(x2 ** 2)                # 値を2乗して正弦(sin)を求める
```

```python
# サンプル4-16-2
# 2×2のグリッドを生成し、ax1～ax4にそれぞれ代入
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
f.suptitle('Sine curve of 4 patterns')
ax1.plot(x1, y1)                     # 1行、1列のAxesにx1の正弦曲線をプロット
ax2.scatter(x2, y2*2, s=10)          # 1行、2列のAxesにx2の正弦×2の分布をプロット
ax3.scatter(x1, y1, color='r', s=10) # 2行、1列のAxesにx1の正弦の分布をプロット
ax4.plot(x2, y2*2, color='r')        # 2行、2列のAxesにｘ2の正弦×2の曲線をプロット

plt.show()
```

```python
# サンプル4-16-3
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(
    2, 2,        # 2×2のグリッド
    sharex=True, # x軸をシェア
    sharey=True  # y軸をシェア
    )
f.suptitle('Sine curve of 4 patterns')
ax1.plot(x1, y1)                     # 1行、1列のAxesにx1の正弦曲線をプロット
ax2.scatter(x2, y2*2, s=10)          # 1行、2列のAxesにx2の正弦×2の分布をプロット
ax3.scatter(x1, y1, color='r', s=10) # 2行、1列のAxesにx1の正弦の分布をプロット
ax4.plot(x2, y2*2, color='r')        # 2行、2列のAxesにｘ2の正弦×2の曲線をプロット

plt.show()
```

```python
# サンプル4-16-4
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(
    2, 2,         # 2×2のグリッド
    sharex='col', # x軸を列方向に対してシェア
    sharey='col'  # y軸を列方向に対してシェア
)
f.suptitle('Sine curve of 4 patterns')
ax1.plot(x1, y1)                     # 1行、1列のAxesにx1の正弦曲線をプロット
ax2.scatter(x2, y2*2, s=10)          # 1行、2列のAxesにx2の正弦×2の分布をプロット
ax3.scatter(x1, y1, color='r', s=10) # 2行、1列のAxesにx1の正弦の分布をプロット
ax4.plot(x2, y2*2, color='r')        # 2行、2列のAxesにｘ2の正弦×2の曲線をプロット

plt.show()
```
