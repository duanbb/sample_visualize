```python
# サンプル4-36-1
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


# 乱数生成時のシード(種)を固定
np.random.seed(10)
# 平均0,標準偏差3,4,5,6,7の正規分布からそれぞれ100個のデータをサンプリング
# サイズ100のリスト×4をdataに格納
data = [np.random.normal(0, std, 100) for std in range(3, 8)]

# Figure上に1行×2列のグリッドを配置
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# バイオリンプロット
axes[0].violinplot(data,
                   showmeans=False, # 平均をマーキングしない
                   showmedians=True # 中央値をマーキングする
                  )
axes[0].set_title('Violin plot')

# ボックスプロット
axes[1].boxplot(data)
axes[1].set_title('Box plot')

# x軸の目盛ラベルを表示
plt.setp(axes,
         # x軸の目盛ラベルの表示位置
         xticks=[y + 1 for y in range(len(data))],
         # 目盛ラベルに表示するテキスト
         xticklabels=['x1', 'x2', 'x3', 'x4', 'x5'])

# グリッド、軸ラベルの表示
for ax in axes:
    ax.yaxis.grid(True)              # y軸のグリッドを表示
    ax.set_xlabel('Four samples')    # x軸ラベル
    ax.set_ylabel('Observed values') # y軸ラベル

plt.show()
```
