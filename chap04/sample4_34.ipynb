```python
# サンプル4-34-1
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# 乱数生成時のシード(種)を固定
np.random.seed(0)

# 平均0,標準偏差1,2,3,4の正規分布からそれぞれ100個のデータをサンプリング
# サイズ100のリスト×4をdataに格納
data = [np.random.normal(0, std, size=100) for std in range(1, 5)]

# x軸のラベルテキストとテキストサイズ
labels =['A', 'B', 'C', 'D']
fs = 10

# Figure上に2行×3列のグリッドを配置
fig, axes = plt.subplots(2, 3, figsize=(10, 10), sharey=True)

# デフォルトのスタイルでボックスプロット
axes[0, 0].boxplot(data,
                   labels=labels    # ラベルをセット
                  )
axes[0, 0].set_title('Default', fontsize=fs)

# 平均をマーキングしてボックスプロット
axes[0, 1].boxplot(data,
                   labels=labels,   # ラベルをセット
                   showmeans=True   # 平均を示す
                  )
axes[0, 1].set_title('showmeans=True', fontsize=fs)

# 平均をラインでマーキングしてボックスプロット
axes[0, 2].boxplot(data,
                   labels=labels,  # ラベルをセット
                   showmeans=True, # 平均を示す
                   meanline=True   # 平均の箇所にラインを引く
                  )
axes[0, 2].set_title('showmeans=True,\nmeanline=True', fontsize=fs)

# ボックス、ウィスカーのキャップを非表示にしてボックスプロット
axes[1, 0].boxplot(data,
                   labels=labels,  # ラベルをセット
                   showbox=False,  # ボックスを非表示
                   showcaps=False  # ウィスカー(ひげ)の先端のキャップを非表示
                  )

# くぼみ付きにしてボックスプロット
axes[1, 0].set_title(
    'Tufte Style \n(showbox=False,\nshowcaps=False)',
    fontsize=fs
)

axes[1, 1].boxplot(data,
                   labels=labels,  # ラベルをセット
                   notch=True,     # ノッチ(くぼみ)付きのボックスにする
                  )                # 1000～10000を設定
axes[1, 1].set_title('notch=True', fontsize=fs)

# ウィスカーの先端を超える外れ値を非表示にしてボックスプロット
axes[1, 2].boxplot(data,
                   labels=labels,  # ラベルをセット
                   showfliers=False# ウィスカーを超える外れ値を表示しない
                  )
axes[1, 2].set_title('showfliers=False', fontsize=fs)

fig.subplots_adjust(hspace=0.3)   # Axesの縦の間隔を空ける

plt.show()
```
