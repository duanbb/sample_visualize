```python
# 4-39-1
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure()             # Figure
ax = axes3d.Axes3D(fig)        # Axes3D

# Prepare arrays x, y, z
theta = np.linspace(-4 * np.pi,# 
                    4 * np.pi, # 
                    100        # 100
                   )
z = np.linspace(-2, 2, 100)    # -22100
r = z**2 + 1
x = r * np.sin(theta)          # thetarx
y = r * np.cos(theta)          # thetary

# Axes3Dplot()
ax.plot(x, y, z, label='parametric curve')
# 
ax.legend()

plt.show()
```
