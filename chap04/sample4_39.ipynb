```python
# サンプル4-39-1
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure()             # Figureを生成
ax = axes3d.Axes3D(fig)        # Axes3Dオブジェクトを配置する

# Prepare arrays x, y, z
theta = np.linspace(-4 * np.pi,# 等差数列の始点
                    4 * np.pi, # 等差数列の終点
                    100        # 100個生成
                   )
z = np.linspace(-2, 2, 100)    # -2から2まで100個生成
r = z**2 + 1
x = r * np.sin(theta)          # thetaの正弦×rをx値
y = r * np.cos(theta)          # thetaの余弦×rをy値

# Axes3Dのplot()でパラメトリックラインをプロット
ax.plot(x, y, z, label='parametric curve')
# 凡例を表示
ax.legend()

plt.show()
```
