```python
# サンプル4-06-1
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 5.0, 0.01)       # 0～5.0の範囲で0.01刻みの等差数列
s = np.cos(2*np.pi*t)               # (2・円周率・t)の余弦(サイン)
plt.plot(t, s, lw=2)                # tをx軸、sをy軸にしてライン幅2でプロット
plt.ylim(-2, 2)                     # y軸のスケールを-2から2にする

plt.annotate(s='max point',         # 注釈用のテキスト(s=はなくてもよい)
             xy=(2, 1),             # 矢印の先端の位置
             xytext=(3, 1.5),       # 注釈テキストの開始位置
             size=14,               # テキストのサイズ
             arrowprops=dict(facecolor='red',   # 矢印を塗りつぶすカラー
                             width=1,           # 矢印の線の部分の太さ
                             headwidth=10,      # 矢印の太さ
                             headlength=15,     # 矢印の長さ
                             linewidth=1,       # エッジラインの太さ
                             edgecolor='black', # エッジラインの色
                             shrink=0.05        # 矢印の先端を若干離す
                            )
             )
plt.show()
```

```python
# サンプル4-06-2
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
plt.plot(t, s, lw=2)
plt.ylim(-2, 2)

plt.text(2.15,                        # テキストの左端：x軸の1.0の位置
         1.05,                        # テキストのベースライン：y軸の5.0の位置
         'max point',                 # 表示するテキスト
         size=12,                     # フォントサイズ20pt
         ha='left',                   # テキストの左端を基準にする
         va='bottom',                 # テキストの下端を基準にする
         rotation= 33,                # 反時計回りに33度回転
         bbox=dict(boxstyle='larrow', # ボックスの形状は左向き矢印
                   facecolor='orange',# ピンクで塗りつぶす
                   edgecolor='red',   # エッジの色は赤
                   linestyle='solid', # エッジラインは実線
                   linewidth=1        # エッジラインの幅は1pt
                   )
         )
plt.show()
```
