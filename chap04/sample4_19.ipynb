```python
# サンプル4-19-1
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.01, 10.0, 0.01)
data1 = np.exp(x) # ネイピア数のx乗
data2 = np.sin(x) # xの正弦
```

```python
# サンプル4-19-2
# Figureにaxを配置
fig, ax = plt.subplots()
ax.plot(x, data1, color='red')   # xとネイピア数のx乗のラインをプロット
ax.plot(x, data2, color='blue')  # xとxの正弦のラインをプロット
ax.set_xlabel('time (s)')        # x軸ラベル

plt.show()
```

```python
# サンプル4-19-3
# Figureにax1を配置
fig, ax1 = plt.subplots()
ax1.plot(x, data1, color='red')              # xとネイピア数のx乗をラインプロット
ax1.set_xlabel('time (s)')                   # x軸ラベル
ax1.set_ylabel('exp', color='red')           # y軸ラベルは赤
ax1.tick_params(axis='y', labelcolor='red')  # y軸の目盛ラベルは赤にする

# 反対側にy軸が配置されたAxesオブジェクトを配置
ax2 = ax1.twinx()
ax2.plot(x, data2, color='blue')             # xとxの正弦をラインプロット
ax2.set_ylabel('sin', color='blue')          # y軸ラベルは青
ax2.tick_params(axis='y', labelcolor='blue') # y軸の目盛ラベルは青

plt.show()
```
