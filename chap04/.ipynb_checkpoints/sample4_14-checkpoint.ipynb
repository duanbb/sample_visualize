```python
# サンプル4-14-1
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 400) # 0～π×2を400等分した等差数列
y = np.sin(x ** 2)                 # 各値を2乗して正弦(sin)を求める
```

```python
# サンプル4-14-2
f, ax = plt.subplots(2, 1,         # 2×1のグリッド
                     sharex=True   # x軸を共有
                    )
f.suptitle('Sharing x- axis')      # Figureのタイトル
ax[0].plot(x, y)                   # 1行目のAxesにxの正弦曲線をプロット
ax[1].scatter(x, y, 10)            # 2行目のAxesにxの正弦の分布をプロット

plt.show()
```

```python
# サンプル4-14-3
f, (ax1, ax2) = plt.subplots(1, 2,       # 1×2のグリッド
                             sharey=True # y軸を共有
                            )
f.suptitle('Sharing y-axis')             # Figureのタイトル
ax1.plot(x, y)                           # 1行、1列のAxesにxの正弦曲線をプロット
ax2.scatter(x, y, 10)                    # 1行、2列のAxesにxの正弦の分布をプロット

plt.show()
```

```python
# サンプル4-14-4
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(
    2, 2,         # 2×2のグリッド
    sharex=True,  # x軸を共有
    sharey=True   # y軸を共有
    )
f.suptitle('Sharing x-axis and y-axis')

# 1行、1列のAxesにx**2の正弦曲線をプロット
ax1.plot(x, y)
# 1行、2列のAxesにx**2の正弦の分布をプロット
ax2.scatter(x, y, s=10)
# 2行、1列のAxesにx**2の正弦を変更してプロット
ax3.scatter(x, 2 * y ** 2 - 1, s=10, color='r')
# 2行、2列のAxesにx**2の正弦を変更した曲線をプロット
ax4.plot(x, 2 * y ** 2 - 1, color='r')

plt.show()
```
