```python
# サンプル4-09-1
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-np.pi, np.pi, 256) # -π～πまでを256等分した等差数列
C,S = np.cos(X), np.sin(X)          # Xの正弦(sin)、余弦(cos)を求める
```

```python
# サンプル4-09-2
_, ax = plt.subplots(figsize=(8, 5))# FigureにAxesを1つ配置
ax.plot(X, C)                       # Xの正弦をプロット
ax.plot(X, S)                       # Xの余弦をプロット

plt.show()
```

```python
# サンプル4-09-3
_, ax = plt.subplots(figsize=(8,5))          # FigureにAxesを1つ配置

ax.spines['right'].set_color('none')         # 右のスパインを消す
ax.spines['top'].set_color('none')           # 上部のスパインを消す
ax.spines['bottom'].set_position(('data',0)) # 下部のスパインをy軸の0の位置へ移動
ax.spines['left'].set_position(('data',0))   # 左側のスパインをy軸の0の位置へ移動
ax.xaxis.set_ticks_position('bottom')        # x軸の目盛ラベルをスパイン(x軸)の下側に表示
ax.yaxis.set_ticks_position('left')          # y軸の目盛ラベルをスパイン(y軸)の右側に表示

ax.plot(X, C,            # Xの正弦をプロット
        color="blue",    # ラインのカラーは青
        linewidth=2.5,   # ラインの太さは2.5pt
        linestyle="-"    # ラインスタイルは実線
       )
ax.plot(X, S,            # Xの余弦をプロット
        color="red",     # ラインのカラーは赤
        linewidth=2.5,   # ラインの太さは2.5pt
        linestyle="-"    # ラインスタイルは実線
       )

ax.set_xlim(X.min()*1.1, # x軸の下限はXの最小値×1.1
            X.max()*1.1  # x軸の上限はXの最大値×1.1
           )

# x軸の目盛ラベルの表示位置
ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
# x軸の目盛ラベル
ax.set_xticklabels(
    [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

# y軸の目盛ラベルの表示位置
ax.set_ylim(C.min()*1.1, 0, C.max()*1.1)
# y軸の目盛ラベル
ax.set_yticks([-1, 0, +1])
ax.set_yticklabels([r'$-1$', r'$0$', r'$+1$'])

plt.show()
```
