```python
# サンプル4-04-1
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(range(10))                   # y値を0～10にしてラインをプロット

plt.text(0,                           # テキストの左端：x軸の0の位置
         8.5,                         # テキストのベースライン：yy軸の8.5の位置
         'test',                      # 表示するテキスト
         size=30                      # フォントサイズ30pt
         )

plt.text(5.0,                         # テキストの左端：x軸の5.0の位置
         5.0,                         # テキストのベースライン：yy軸の5.0の位置
         'test',                      # 表示するテキスト
         size=30,                     # フォントサイズ30pt
         bbox=dict(boxstyle='circle', # ボックスの形状は円
                   facecolor='pink',  # ピンクで塗りつぶす
                   edgecolor='red',   # エッジの色は赤
                   linestyle='solid', # エッジラインは実線
                   linewidth=5        # エッジラインの幅は5pt
                   )
         )

plt.text(1.0,                         # テキストの左端：x軸の1.0の位置
         5.0,                         # テキストのベースライン：y軸の5.0の位置
         'comment',                   # 表示するテキスト
         size=20,                     # フォントサイズ20pt
         rotation= 33,                # 反時計回りに33度回転
         bbox=dict(boxstyle='round',  # ボックスの形状は角丸四角形
                   facecolor='orange',# ピンクで塗りつぶす
                   edgecolor='red',   # エッジの色は赤
                   linestyle='solid', # エッジラインは実線
                   linewidth=1        # エッジラインの幅は1pt
                   )
         )

plt.show()
```

```python
# サンプル4-04-2
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4, 5, 6],                      # xの値
         [8.6, 5.3, 10.2, 16.1, 22.3, 24.6],      # yの値
         marker='o',                              # サークル型のマーカー
         label='Average Temperature\n(Celsius)'   # 凡例のラベルを2行にする
        )
plt.legend()                                      # 凡例をロット

plt.title('Average Temperature\n2018 first half', # タイトルを2行にする
          size=18)

plt.xlabel('Month\n(2018)',                       # x軸のラベルを2行にする
           size=14,            
           bbox=dict(boxstyle='square',           # 四角形のボックス     
                     facecolor='none'             # ボックスの塗りつぶしなし
                     )
          )
plt.ylabel('Temperature\n(from 0 to 25 degrees)', # y軸のラベルを2行にする
           size=14,
           bbox=dict(boxstyle='square',           # 四角形のボックス 
                     facecolor='none'             # ボックスの塗りつぶしなし
                     )
           )

plt.xticks([1, 2, 3, 4, 5, 6],                    # 目盛ラベルを配置するx軸の位置
           ['Jan.\n2018', 'Feb.\n2018',           # xの目盛ラベル
            'Mar.\n2018', 'Apr.\n2018',
            'May\n2018', 'Jun.\n2018'],
           size=14
          )

plt.yticks([0, 5, 10, 15, 20, 25],                # 目盛ラベルを配置するy軸の位置
           ['0℃', '5℃', '10℃',                 # yの目盛ラベル
            '15℃', '20℃', '25℃'],
           size=12)

plt.text(6.5,                        # x軸の6.5(エリア外)の位置に配置
         25,                         # y軸の25の位置に配置
         'Every day\n at 12 oclock\n measurement',
         size=18,
         ha='left',                  # テキストの左端を基準にする
         va='top',                   # テキストの上端を基準にする
         multialignment="center",    # テキストを中央揃えで配置
         bbox=dict(boxstyle='round', # 四角形のボックス
                   facecolor='none'  # ボックスの塗りつぶしなし
                   )
        )

plt.show()
```
