```python
# サンプル4-27-1
%matplotlib inline
import matplotlib.pyplot as plt

labels = 'Apple', 'Banana', 'Grape', 'Orange' # グラフ要素のラベル
sizes = [15, 30, 45, 10] # グラフ要素の値
explode = [0, 0, 0.1, 0] # 各要素の中心位置

fig, ax = plt.subplots() # Figure上にAxesを配置
ax.pie(sizes,            # グラフ要素の値を設定
       explode=explode,  # 3番目の要素の中心位置を0.1だけ離す
       labels=labels,    # グラフ要素のラベルを設定
       autopct='%.1f%%', # 構成割合として小数点以下2桁までをプロット
       shadow=True,      # グラフ要素に影を付ける
       startangle=90     #
      )
ax.axis('equal')         # グラフを真円仁する

plt.show()
```
