```python
# サンプル4-18-1
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter

# 静菌平均0,標準偏差1の標準正規分布から1000個のデータをサンプリング
y = np.random.normal(loc=0, scale=1, size=1000)
# 0と1を除く
y = y[(y > 0) & (y < 1)]
# 昇順に並べ替える
y.sort()
# 0～yの値の数の範囲で1刻みの等差数列を生成
x = np.arange(len(y))

# 2×2のグリッドを生成し、x軸を共有する
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2,  figsize=(8, 8), sharex=True)

# linear(1次関数)のグラフ
ax1.plot(x, y)
ax1.set_yscale('linear')
ax1.set_title('linear')
ax1.grid(True)

# log(対数)のグラフ
ax2.plot(x, y)
ax2.set_yscale('log')
ax2.set_title('log')
ax2.grid(True)

# symmetric log（対数関数の対称）のグラフ
ax3.plot(x, y - y.mean())      # x,y-平均値のラインをプロット
ax3.set_yscale('symlog',
              linthreshy=0.02 # （-x、x）の範囲を指定する
             )
ax3.set_title('symmetric log')
ax3.grid(True)

# logit(ロジット関数)のグラフ
ax4.plot(x, y)
ax4.set_yscale('logit')
ax4.set_title('logit')
ax4.grid(True)
# y軸ラベルのための処理
ax4.yaxis.set_minor_formatter(NullFormatter())

plt.show()
```
