```python
# サンプル4-26-1
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Appleの1日当たりの平均売上と標準偏差
apple_means, apple_std = (35, 25, 30, 35, 22), (2, 3, 4, 5, 2)
# Orangeの1日当たりの平均売上と標準偏差
orange_means, orange_std = (38, 28, 34, 28, 25), (3, 5, 2, 3, 3)

# 0からfruitの数だけ等差数列を生成し、バーの位置の基準にするx値とする
x_pos = np.arange(len(apple_means))
# バーの幅
width = 0.35
# Figure上にAxesを配置
fig, ax = plt.subplots(figsize=(8, 8))

# 左に配置するバー
rects1 = ax.bar(x_pos - width,  # バーの左端をwidthの分だけ左にずらす
                apple_means,    # appleの月別の1日当たりの平均売上
                width,          # バーの幅 
                align='edge',   # バーの左端を基準に配置
                yerr=apple_std, # 標準偏差を誤差としてエラーバーを表示
                color='blue',   # バーの色
                ecolor='black', # エラーバーの色  
                label='Apple'   # 凡例のラベル
               )
# 右に配置するバー
rects2 = ax.bar(x_pos,           # バーの左端をx_posの位置にする
                orange_means,    # orangeの月別の1日当たりの平均売上
                width,           # バーの幅
                align='edge',    # バーの左端を基準に配置
                yerr=orange_std, # 標準偏差を誤差としてエラーバーを表示
                color='red',     # エラーバーの色
                label='Orange'   # 凡例のラベル
               )

# タイトル、軸ラベル、目盛ラベル、凡例の表示
ax.set_title('Average sales per day') # タイトル
ax.legend()                           # 凡例を表示
ax.set_ylabel('Salse')                # y軸ラベル
ax.set_xticks(x_pos)                  # x軸の目盛ラベルの位置
# 目盛ラベルのテキスト
ax.set_xticklabels(('Jan.', 'Feb.', 'Mar.', 'Apr.', 'May'))

def set_label(rects, xpos):
    """
    バーの高さを表す数値をバーの先端に表示
    表示位置として左バーはエラーバーの左、右バーはエラーバーの右にする
    
    パラメーター
    rects：Rectangleオブジェクト
    xpos：テキストの水平方向の基準値、'center' 'right' 'left'のどれか
    """
    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 0.5, 'right': 0.57, 'left': 0.43}

    for rect in rects: # BarContainerからバーの情報を取り出す
        height = rect.get_height() # バーの高さを取得
        ax.text(
            # x軸上の位置
            # xの位置を取得し、バーの幅にオフセット値を掛けた値を合算する
            rect.get_x() + rect.get_width()*offset[xpos], # x軸の位置
            # y軸上の位置は高さ×1.01
            1.01*height,
            # バーの高さをテキストとして出力
            '{}'.format(height),
            # テキストの水平方向の基準を'center' 'right' 'left'のどれかに設定
            ha=ha[xpos],
            # テキストの垂直方向の基準をテキスト下端に設定
            va='bottom'
        )

set_label(rects1, 'left')
set_label(rects2, 'right')

plt.show()
```
