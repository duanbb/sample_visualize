```python
# サンプル4-28-1
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.transforms as mtransforms
import numpy as np

from matplotlib.transforms import offset_copy

xs = np.arange(8)                    # 0～7の1刻みの等差数列(8個)
ys = xs**2                           # xsの要素を2乗する

fig = plt.figure(figsize=(10, 15))   # Figureを生成

# 1行目のグリッドにAxesを追加
ax1 = fig.add_subplot(211)

# 現在のTransformオブジェクトオフセットのコピーを取得し、
# x,yのオフセットを変換する
trans_offset = mtransforms.offset_copy(
    ax1.transData, # 変換対象のAxesのTransformオブジェクト
    fig=fig,       # 対象のAxesが配置されているFigureオブジェクト
    x=0.05,        # xのオフセット
    y=0.10,        # yのオフセット
    units='inches' # offsetの単位をインチにする
)

# マーカーとラベルテキストの描画
for x, y in zip(xs, ys):
    ax1.plot(x, y, 'ro')                  # マーカーをプロット
    ax1.text(x, y,                        # 描画位置の座標
             '%d, %d' % (int(x), int(y)), # x値,y値を表示
             transform=trans_offset,      # 変更後のオフセットを適用する
             horizontalalignment='left',  # 水平方向の基準をテキスト左端にする
             verticalalignment='bottom'   # 垂直方向の基準をテキスト下端にする
            )

# 1行目のグリッドに極座標グラフ用のAxesを追加
ax2 = fig.add_subplot(212,                # 2行目のグリッドにAxesを追加
                      projection='polar'  # 極座標グラフにする
                     )

# 現在のTransformオブジェクトオフセットのコピーを取得し、
# x,yのオフセットを変換する
trans_offset = mtransforms.offset_copy(
    ax2.transData,
    fig=fig,
    y=6,
    units='dots'
)

# マーカーとラベルテキストの描画
for x, y in zip(xs, ys):
    ax2.plot(x, y, 'ro')                   # マーカーをプロット
    ax2.text(x, y,                         # 描画位置の座標
             '%d, %d' % (int(x), int(y)),  # x値,y値を表示
             transform=trans_offset,       # 変更後のオフセットを適用する
             horizontalalignment='center', # 水平方向の基準をテキスト中央にする
             verticalalignment='bottom'    # 垂直方向の基準をテキスト下端にする
            )

plt.show()
```
