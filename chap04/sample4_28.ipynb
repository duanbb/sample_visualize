```python
# 4-28-1
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.transforms as mtransforms
import numpy as np

from matplotlib.transforms import offset_copy

xs = np.arange(8)                    # 071(8)
ys = xs**2                           # xs2

fig = plt.figure(figsize=(10, 15))   # Figure

# 1Axes
ax1 = fig.add_subplot(211)

# Transform
# x,y
trans_offset = mtransforms.offset_copy(
    ax1.transData, # AxesTransform
    fig=fig,       # AxesFigure
    x=0.05,        # x
    y=0.10,        # y
    units='inches' # offset
)

# 
for x, y in zip(xs, ys):
    ax1.plot(x, y, 'ro')                  # 
    ax1.text(x, y,                        # 
             '%d, %d' % (int(x), int(y)), # x,y
             transform=trans_offset,      # 
             horizontalalignment='left',  # 
             verticalalignment='bottom'   # 
            )

# 1Axes
ax2 = fig.add_subplot(212,                # 2Axes
                      projection='polar'  # 
                     )

# Transform
# x,y
trans_offset = mtransforms.offset_copy(
    ax2.transData,
    fig=fig,
    y=6,
    units='dots'
)

# 
for x, y in zip(xs, ys):
    ax2.plot(x, y, 'ro')                   # 
    ax2.text(x, y,                         # 
             '%d, %d' % (int(x), int(y)),  # x,y
             transform=trans_offset,       # 
             horizontalalignment='center', # 
             verticalalignment='bottom'    # 
            )

plt.show()
```
